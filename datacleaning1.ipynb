```python
df["Rotten Tomatoes"] = df["Rotten Tomatoes"].str.split(pat="/", expand=True)[0]
df["Rotten Tomatoes"] =df["Rotten Tomatoes"].str.replace("/100","").astype(int)
```

```python
df["IMDb"] = df["IMDb"].str.split(pat="/", expand=True)[0]
df["IMDb"] =df["IMDb"].str.replace("/10","").astype(float)
```

```python
df.dtypes
```

```python
def replace_IMDb(row):
  if pd.isnull(row["IMDb"])==True:
    rating=row["Rotten Tomatoes"]/10
    row["IMDb"]=rating
  return row

df=df.apply(lambda row: replace_IMDb(row), axis=1)
```
